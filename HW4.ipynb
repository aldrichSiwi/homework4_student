  >a) Find the indices of bad data quality values in `data_flag` by numpy `where()` and store in `bad_inds`.

  >b) Save the corresponding "year" and "depth_mm" values of bad data in a separate arrays, `bad_years` and `bad_depths` respectively.

1. A) The mask value of the dataset is -99999. This value is used to indicate missing sea level measurements in the dataset.

2. A) The delimiter type for this text file is semicolon (;). This is used to separate the columns of data, such as the "Year and decimal month", "Depth (mm)", and other attributes in the dataset.

````python
# import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate 
from scipy import stats
print("Part 1")
# your code

# answer variables 
filepath = "Pakistan_sea_level_psmsl.txt"
# opening the file
with open(filepath, "r") as fileobj:
    header_lines = 19
    # for loop
    for i in range(header_lines):
        header = fileobj.readline().strip()
        
display(header_lines)
print("Part 2")
# your code

# answer variables
# read each column 
sea_level = np.genfromtxt(filepath, delimiter=";", skip_header=header_lines, dtype=float)

# Extract specific columns into separate arrays
year_month = sea_level[:, 0]
depth_mm = sea_level[:, 1]
data_flag = sea_level[:, 3]

#display
display(sea_level)
display(year_month)
display(depth_mm)
display(data_flag)

print("Part 3")
# your code
# mask the value
mask_value = -99999
# answer variables
# Find indices of masked depth values
mask_inds = np.where(depth_mm == mask_value)[0]
# replace it with np.nan
depth_mm[mask_inds] = np.nan
# display it
display(mask_inds)
display(depth_mm)

print("Part 4")
# your code
# value indicating bad data
bad_flag = 11

# answer variables
#  Find indices of bad data quality values
bad_inds = np.where(data_flag == bad_flag)[0]

# Save corresponding "year" and "depth_mm" values of bad data
badyears = year_month[bad_inds]
baddepths = depth_mm[bad_inds]

# Replace bad depth values in depth_mm with np.nan
depth_mm[bad_inds] = np.nan
# print result
print(bad_inds)
print(badyears)
print(baddepths)
print(depth_mm)
````
